```python
import numpy as np
import os.path as op
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from hippocampus.plotting import tsplot_boot
from definitions import RESULTS_FOLDER
import matplotlib

sns.palplot(sns.color_palette())
```

```python
results_dir = op.join(RESULTS_FOLDER, 'pearce', 'lesion_DLS')
```

```python
ETs = []
for a in range(100):
    df = pd.read_csv(op.join(results_dir, 'agent{}.csv'.format(a)))
    escape_time = df.pivot_table(values='escape time', index=['session', 'trial'])
    escape_time['agent'] = a
    ETs.append(escape_time)
```

```python
len(ETs)
```

```python
df=pd.concat(ETs).reset_index()
```

```python
df = df.set_index(['agent', 'session', 'trial']).reset_index()
df = df[np.logical_or(df.trial==0,df.trial==3)]
df.session = df.session + 1
```

```python
pal = [sns.color_palette()[2], sns.color_palette()[2]]

font = {'size'   : 18}

matplotlib.rc('font', **font)

fig, ax = plt.subplots(figsize=(6,5))
sns.lineplot(x='session', y='escape time', data=df, ax=ax, 
             ci=None, linewidth=3, palette=pal, style='trial')
plt.legend(['DLS lesion - trial 1','DLS lesion - trial 2'])

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.ylabel('Escape time')
plt.xlabel('Session')

plt.xticks(range(1,12))
plt.tight_layout()
plt.savefig(op.join(RESULTS_FOLDER, 'figures', 'DLS_lesion_watermaze.pdf'))
```
